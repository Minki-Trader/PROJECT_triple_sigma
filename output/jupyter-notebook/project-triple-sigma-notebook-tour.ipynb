```python
# Setup cell: use only the standard library so the notebook stays easy to run.
from __future__ import annotations

from collections import Counter
from pathlib import Path
import csv

def find_project_root(start: Path) -> Path:
    for candidate in [start, *start.parents]:
        if (candidate / "_coord").exists() and (candidate / "design").exists():
            return candidate
    raise FileNotFoundError("Could not find PROJECT_triple_sigma root from the current working directory.")

PROJECT_ROOT = find_project_root(Path.cwd())
PROJECT_ROOT
```

## Step 1 - Pick one retained trade log

Start with a known small artifact so the first run is predictable.
Later you can change the path to `step20_live_be_probe`,
`step20_live_pass_regression`, or another retained folder.


```python
SAMPLE_TRADE_LOG = PROJECT_ROOT / "_coord" / "artifacts" / "step20_control_trade" / "trade_log.csv"
SAMPLE_TRADE_LOG.exists(), SAMPLE_TRADE_LOG
```

## Step 2 - Read the CSV into memory

This cell loads the file with `csv.DictReader`, then shows the row count
and the first two rows so you can see the column names and values.


```python
with SAMPLE_TRADE_LOG.open("r", encoding="utf-8-sig", newline="") as f:
    rows = list(csv.DictReader(f))

len(rows), rows[:2]
```

## Step 3 - Build a small summary

Here we count `event_type` values and then count `exit_reason` only for
`EXIT` rows. This is the kind of small summary that notebooks are very
good at when reviewing retained artifacts.


```python
event_counts = Counter(row["event_type"] for row in rows)
exit_reason_counts = Counter(
    row["exit_reason"]
    for row in rows
    if row["event_type"] == "EXIT"
)

{
    "event_counts": dict(event_counts),
    "top_exit_reasons": dict(exit_reason_counts.most_common(10)),
}
```
